```python
import tensorflow as tf

fashion_mnist = tf.keras.datasets.fashion_mnist
(train_X, train_Y),(test_X, test_Y) = fashion_mnist.load_data()

train_X = train_X / 255
test_X = test_X / 255

train_X = train_X.reshape(-1, 28, 28, 1)
test_X = test_X.reshape(-1, 28, 28, 1)

print(f"train_X의 shape은 {train_X.shape}이고 train_Y의 shape은 {train_Y.shape}이다.")
print(f"test_X의 shape은 {test_X.shape}이고 test_Y의 shape은 {test_Y.shape}이다.")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.imshow(train_X[1])

plt.subplot(1,2,2)
plt.imshow(test_X[1])
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

image_generator = ImageDataGenerator(rotation_range=10, shear_range=0.5, zoom_range=0.1,\
    width_shift_range=0.10, height_shift_range=0.10, horizontal_flip=True, vertical_flip=False)

augument_size =  100

x_augumented = image_generator.flow(
    x = np.tile(A=train_X[1].reshape(28*28), reps = 100).reshape(-1, 28, 28, 1),\
    y = np.zeros(augument_size), batch_size=augument_size, shuffle=False).next()[0]
            
```

```python
plt.figure(figsize=(16, 16))

for c in range(augument_size):
    plt.subplot(10, 10, c+1)
    plt.axis('off')
    plt.imshow(x_augumented[c].reshape(28, 28), cmap='gray')
plt.show()
```
